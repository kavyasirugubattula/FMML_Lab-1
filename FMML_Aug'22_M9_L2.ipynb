# Data preparation

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the iris dataset
iris = load_iris()
X = iris['data']
y = iris['target']
names = iris['target_names']
feature_names = iris['feature_names']

# Scale data to have mean 0 and variance 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data set into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=2)
```

# Visualising dataset

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
for target, target_name in enumerate(names):
    X_plot = X[y == target]
    ax1.plot(X_plot[:, 0], X_plot[:, 1],
             linestyle='none',
             marker='o',
             label=target_name)
ax1.set_xlabel(feature_names[0])
ax1.set_ylabel(feature_names[1])
ax1.axis('equal')
ax1.legend();

for target, target_name in enumerate(names):
    X_plot = X[y == target]
    ax2.plot(X_plot[:, 2], X_plot[:, 3],
             linestyle='none',
             marker='o',
             label=target_name)
ax2.set_xlabel(feature_names[2])
ax2.set_ylabel(feature_names[3])
ax2.axis('equal')
ax2.legend();
```

# MLP for multiclass classification

```python
import torch
import torch.nn.functional as F
import torch.nn as nn
from torch.autograd import Variable
import tqdm
```

```python
# Defining the model architecture
class Model(nn.Module):
    def __init__(self, input_dim):
        super(Model, self).__init__()
        self.layer1 = nn.Linear(input_dim, 50)
        self.layer2 = nn.Linear(50, 50)
        self.layer3 = nn.Linear(50, 3)

    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        x = F.softmax(self.layer3(x), dim=1)
        return x
```

```python
# Instantiating the model, using Adam optimiser, and Cross Entropy Loss, which is quite commonlu used for classification tasks.
model     = Model(X_train.shape[1])
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn   = nn.CrossEntropyLoss()
```

# Train the model

```python
# Train for 100 epochs
EPOCHS  = 100
X_train = torch.from_numpy(X_train).float()
X_test= torch.from_numpy(X_test).float()
y_test = torch.from_numpy(y_test)
y_train = torch.from_numpy(y_train)


loss_list     = np.zeros((EPOCHS,))
accuracy_list = np.zeros((EPOCHS,))

for epoch in tqdm.trange(EPOCHS):
    y_pred = model(X_train)
    loss = loss_fn(y_pred, y_train)
    loss_list[epoch] = loss.item()

    # Zero gradients
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    with torch.no_grad():
        y_pred = model(X_test)
        correct = (torch.argmax(y_pred, dim=1) == y_test).type(torch.FloatTensor)
        accuracy_list[epoch] = correct.mean()
```

# Plot training progress

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(12, 6), sharex=True)

ax1.plot(accuracy_list)
ax1.set_ylabel("validation accuracy")
ax2.plot(loss_list)
ax2.set_ylabel("validation loss")
ax2.set_xlabel("epochs");
```

##**QUESTION-1**

For instance, in the case of XOR, which is a classic example of a problem that cannot be solved with a single-layer perceptron (without nonlinear activation functions), you can successfully solve it using a neural network with one or more hidden layers.

Here's how changing the number of hidden layers might affect the performance:

**No Hidden Layers (Single-Layer Perceptron):**
With no hidden layers, the network behaves like a logistic regression model, unable to learn the nonlinear XOR function. It can only separate the data linearly, leading to poor performance on the XOR problem.

**One Hidden Layer:**
Adding one hidden layer with nonlinear activation functions (such as ReLU or sigmoid) allows the network to learn nonlinear decision boundaries. It can effectively solve the XOR problem and other moderately complex classification tasks.

**Multiple Hidden Layers:**
Adding more hidden layers further increases the network's capacity to learn complex patterns in the data. However, adding too many layers can lead to overfitting, where the network learns to memorize the training data rather than generalize to unseen examples.

In summary, adjusting the number of hidden layers in an MLP can enable it to learn more complex relationships in the data, but finding the optimal architecture requires balancing model complexity with the risk of overfitting.








##**QUESTION-2**

Certainly! Altering the number of neurons in the hidden layer of a neural network can also have a significant impact on its performance and capacity to learn from the data. Here's how changing the number of neurons in the hidden layer might affect the network:

**Fewer Neurons:**
If you decrease the number of neurons in the hidden layer, the model's capacity to learn complex patterns in the data decreases. This reduction in capacity might result in underfitting, where the model fails to capture important patterns in the data, leading to poor performance, especially for more complex tasks.

**More Neurons:**
Increasing the number of neurons in the hidden layer can enhance the model's capacity to learn intricate relationships in the data. This increased capacity may enable the model to better fit the training data, potentially improving performance. However, adding too many neurons can also increase the risk of overfitting, where the model memorizes the training data without generalizing well to unseen data.

Finding the optimal number of neurons in the hidden layer often involves experimentation and balancing between model complexity and generalization ability. It depends on factors such as the complexity of the task, the amount of available training data, and the desired trade-off between bias and variance in the model.
